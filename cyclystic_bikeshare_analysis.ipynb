```python
## Aggregating all of the monthly data sets into one dataframe
import pandas as pd
df1 = pd.read_csv('raw_data/202504-divvy-tripdata.csv')
df2 = pd.read_csv('raw_data/202505-divvy-tripdata.csv')
df3 = pd.read_csv('raw_data/202506-divvy-tripdata.csv')
df4 = pd.read_csv('raw_data/202507-divvy-tripdata.csv')
df5 = pd.read_csv('raw_data/202508-divvy-tripdata.csv')
df6 = pd.read_csv('raw_data/202509-divvy-tripdata.csv')
df7 = pd.read_csv('raw_data/202510-divvy-tripdata.csv')
df8 = pd.read_csv('raw_data/202511-divvy-tripdata.csv')
df9 = pd.read_csv('raw_data/202512-divvy-tripdata.csv')
df10 = pd.read_csv('raw_data/202601-divvy-tripdata.csv')
df11 = pd.read_csv('raw_data/202602-divvy-tripdata.csv')
df12 = pd.read_csv('raw_data/202603-divvy-tripdata.csv')

df = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12], ignore_index=True)

## Checking to make sure the size of the dataframe and the variable types are consistent and make sense
print(df.shape)
print(df.dtypes)
```

```python
## Getting a quick look at what the first few rows of data look like and names of the variables
print(df.columns.tolist())
print(df.head())
```

```python
## Checking for any duplicate rows
print(df.duplicated().sum())
```

```python
## Checking for any null values
print(df.isnull().sum())
```

```python
## Check to see if all null values in the station columns are from electric bikes
print(df[df['start_station_name'].isnull()]['rideable_type'].value_counts())
```

```python
## Fill in null values with a placeholder
df['start_station_name'] = df['start_station_name'].fillna('No Station - Electric Bike')
df['start_station_id'] = df['start_station_id'].fillna('No Station - Electric Bike')
df['end_station_name'] = df['end_station_name'].fillna('No Station - Electric Bike')
df['end_station_id'] = df['end_station_id'].fillna('No Station - Electric Bike')

## Verify the nulls have been filled
print(df.isnull().sum())
```

```python
## Delete rows where end latitude and longitude are missing values
df_clean = df.dropna(subset=['end_lat', 'end_lng']).copy()
print(df_clean.shape)

## Verify they are gone
print(df_clean.isnull().sum())

## 5,784 rows with null end coordinates were dropped from the set, repesenting less than 0.1% of the data.
## Likely a result of incomplete trips or a GPS failure making these records useless for this analysis.
```

```python
print(df_clean['started_at'].head(20))
print(df_clean['started_at'].dtype)
```

```python
## Fix some formatting errors in the started_at and ended_at columns
df_clean['started_at'] = pd.to_datetime(df_clean['started_at'], format = 'mixed', errors = 'coerce')
df_clean['ended_at'] = pd.to_datetime(df_clean['ended_at'], format = 'mixed', errors = 'coerce')
df_clean = df_clean.dropna(subset = ['started_at', 'ended_at'])
print(df_clean.shape)
```

```python
## Calculate new variables:

## Ride length
df_clean['ride_length'] = (pd.to_datetime(df_clean['ended_at']) - 
                           pd.to_datetime(df_clean['started_at'])).dt.total_seconds() / 60
## Day of the week
df_clean['day_of_week'] = pd.to_datetime(df_clean['started_at']).dt.day_name()

## Month
df_clean['month'] = pd.to_datetime(df_clean['started_at']).dt.month_name()

## Hour of day
df_clean['hour'] = pd.to_datetime(df_clean['started_at']).dt.hour
```

```python
## Checking to make sure the calculations look right and deleting and zero or negative ride time rows
print(df_clean[['started_at', 'ended_at', 'ride_length', 'day_of_week', 'month', 'hour']].head(10))
print(df_clean[df_clean['ride_length'] <= 0].shape)
df_clean = df_clean[df_clean['ride_length'] > 0]
print(df_clean.shape)

## 29 rows with negative ride times were deleted, a data entry error where the end time was before the start time
```

```python
## Finding mean and median ride times for riders by member status and finding how many riders were casual or members
print(df_clean.groupby('member_casual')['ride_length'].mean())
print(df_clean.groupby('member_casual')['ride_length'].median())
print(df_clean['member_casual'].value_counts())
```

```python
## Find ride counts by rider type per day
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df_clean['day_of_week'] = pd.Categorical(df_clean['day_of_week'], categories=day_order, ordered=True)
print(df_clean.groupby(['member_casual', 'day_of_week'], observed = True)['ride_id'].count())
```

```python
## Find ride counts by rider type per hour of the day
print(df_clean.groupby(['member_casual', 'hour'])['ride_id'].count())
```

```python
## Set Categorical order
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

## Find ride counts by rider type per month
df_clean['month'] = pd.Categorical(df_clean['month'], categories=month_order, ordered=True)
print(df_clean.groupby(['member_casual', 'month'], observed=True)['ride_id'].count())
```

```python
## Find bike type counts by type of rider
print(df_clean.groupby(['member_casual', 'rideable_type'])['ride_id'].count())
```

```python
## Find average length or ride by bike type and rider type
print(df_clean.groupby(['member_casual', 'rideable_type'])['ride_length'].mean())
```

```python
## Find top starting stations
print(df_clean[df_clean['member_casual'] == 'casual'].groupby('start_station_name')['ride_id'].count().sort_values(ascending=False).head(10))
print(df_clean[df_clean['member_casual'] == 'member'].groupby('start_station_name')['ride_id'].count().sort_values(ascending=False).head(10))
```

```python
## Save dataframe as a csv file
df_clean.to_csv('cyclistic_cleaned.csv', index=False)
```
